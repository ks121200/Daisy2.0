```python
import os
import cv2
import numpy as np

image_data_folder = '/content/drive/My Drive/Image_Data'
threshold = 70  # You can adjust this threshold value

good_data_folder = f'/content/drive/My Drive/Good_Data_{threshold}'
blurred_data_folder = f'/content/drive/My Drive/Blurred_Data_{threshold}'

os.makedirs(good_data_folder, exist_ok=True)
os.makedirs(blurred_data_folder, exist_ok=True)
```

## Filtering Data

```python
def is_blurred(image_path, threshold):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Calculate the contrast by taking the absolute difference between the pixel and its neighbors
    contrast = np.abs(img[:-1, :-1].astype(np.float32) - img[1:, 1:].astype(np.float32))
    mean_contrast = np.mean(contrast)

    return mean_contrast < threshold

# Loop through the images in the image_data_folder and move them to the appropriate folders
for image_filename in os.listdir(image_data_folder):
    if image_filename.endswith(('.jpg', '.jpeg', '.png','.tif')):
        image_path = os.path.join(image_data_folder, image_filename)
        if is_blurred(image_path, threshold):
            destination_path = os.path.join(blurred_data_folder, image_filename)
        else:
            destination_path = os.path.join(good_data_folder, image_filename)
        os.rename(image_path, destination_path)
```

## Add source directory

```python
image_data_folder = '/content/drive/MyDrive/Image_Data'
blurred_folder = '/content/drive/MyDrive/Blurred_Data'
good_folder = '/content/drive/MyDrive/Good_Data'

os.makedirs(blurred_folder, exist_ok=True)
os.makedirs(good_folder, exist_ok=True)
```

## Filtering

```python
import cv2
import numpy as np

def is_blurred(image_path, threshold=400):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Perform FFT on the image
    f_transform = np.fft.fft2(img)
    f_transform_shifted = np.fft.fftshift(f_transform)
    magnitude_spectrum = np.log(np.abs(f_transform_shifted) + 1)

    # Calculate the mean magnitude in the frequency domain
    mean_magnitude = np.mean(magnitude_spectrum)

    return mean_magnitude < threshold

# Loop through the images in the image_data_folder and move them to the appropriate folders
for image_filename in os.listdir(image_data_folder):
    if image_filename.endswith(('.jpg', '.jpeg', '.png','.tif')):
        image_path = os.path.join(image_data_folder, image_filename)
        if is_blurred(image_path):
            destination_path = os.path.join(blurred_folder, image_filename)
        else:
            destination_path = os.path.join(good_folder, image_filename)
        os.rename(image_path, destination_path)
```

## DATA AUGMENTATION

```python
import os
from PIL import Image
from skimage import img_as_float, img_as_ubyte
from skimage.filters import unsharp_mask
from skimage.exposure import equalize_adapthist
import cv2
import numpy as np

# Define the folder paths
input_folder = '/content/drive/MyDrive/Image_Data'
output_folder = '/content/drive/MyDrive/Image_Cropped_2'

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# List all files in the input folder
file_list = os.listdir(input_folder)

# Define the percentage of the image to crop
remove_percentage = 0.10  # 90%

# Parameters for bilateral filter
sigma_spatial = 15
sigma_color = 0.1

# Crop the lower 10% of each image and apply Adaptive Histogram Equalization, unsharp masking, and bilateral filter
for file_name in file_list:
    if file_name.endswith(('.tif')):
        # Open the image
        image_path = os.path.join(input_folder, file_name)
        image = Image.open(image_path)

        # Get image dimensions
        width, height = image.size

        # Calculate the height to keep (top 90%)
        keep_height = int(height * (1 - remove_percentage))

        # Calculate the cropping region (top 90%)
        crop_region = (0, 0, width, keep_height)

        # Crop the image
        cropped_image = image.crop(crop_region)

        # Convert the cropped image to a floating-point data type
        cropped_image_float = img_as_float(cropped_image)

        # Ensure the image has three color channels
        if cropped_image_float.shape[-1] != 3:
            cropped_image_float = np.stack([cropped_image_float] * 3, axis=-1)

        # Apply unsharp masking to the cropped image
        sharpened_image = unsharp_mask(cropped_image_float, radius=2, amount=2)

        # Apply Adaptive Histogram Equalization
        equalized_image = equalize_adapthist(sharpened_image)

        # Convert the image to uint8 format for OpenCV
        equalized_image_cv2 = img_as_ubyte(equalized_image)

        # Apply bilateral filter using OpenCV
        filtered_image_cv2 = cv2.bilateralFilter(equalized_image_cv2, -1, sigma_color, sigma_spatial)

        # Convert the filtered image back to float format
        filtered_image_float = img_as_float(filtered_image_cv2)

        # Ensure the image has three color channels
        if filtered_image_float.shape[-1] != 3:
            filtered_image_float = np.stack([filtered_image_float] * 3, axis=-1)

        # Convert the filtered image to 'L' mode (grayscale, compatible with PNG)
        filtered_image_pil = Image.fromarray((filtered_image_float * 255).astype(np.uint8))

        # Define the filename for the filtered image (in PNG format)
        filtered_image_filename = os.path.join(output_folder, f'filtered_{file_name[:-4]}.png')

        # Save the cropped, sharpened, equalized, and filtered image back to the folder in PNG format
        filtered_image_pil.save(filtered_image_filename)
```

```python
from google.colab import drive, files
import cv2
import os
import numpy as np

# Mount Google Drive
drive.mount('/content/drive')

# Specify the folder names
input_folder_name = 'image_cropped'
output_folder_name = 'image_CNN'

# Create output folder if it doesn't exist
output_folder_path = os.path.join('/content/drive/My Drive', output_folder_name)
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

# Set the path to the input folder in Google Drive
input_folder_path = os.path.join('/content/drive/My Drive', input_folder_name)

# Process each image in the folder
for filename in os.listdir(input_folder_path):
    image_path = os.path.join(input_folder_path, filename)

    # Load the image
    image = cv2.imread(image_path)

    # Check if the image is loaded successfully
    if image is None or image.size == 0:
        print(f"Error loading image: {image_path}")
        continue

    # Sanitize the image name (replace spaces with underscores)
    image_name = os.path.splitext(filename)[0].replace(" ", "_")

    # Save the original image
    original_output_path = os.path.join(output_folder_path, f'{image_name}_original.jpg')
    print(f"Attempting to write original image to: {original_output_path}")
    cv2.imwrite(original_output_path, image)
    files.download(original_output_path)  # Download the original image

    # Rotate and save copies using NumPy's rotation function
    for angle in [90, 180, 270]:
        rotated_image = np.rot90(image, k=angle // 90)

        rotated_output_path = os.path.join(output_folder_path, f'{image_name}_rotate{angle}.jpg')
        print(f"Attempting to write rotated image to: {rotated_output_path}")
        cv2.imwrite(rotated_output_path, rotated_image)
        files.download(rotated_output_path)  # Download the rotated image

    # Mirror and save copies
    mirrored_image = cv2.flip(image, 1)
    mirrored_output_path = os.path.join(output_folder_path, f'{image_name}_mirrored.jpg')
    print(f"Attempting to write mirrored image to: {mirrored_output_path}")
    cv2.imwrite(mirrored_output_path, mirrored_image)
    files.download(mirrored_output_path)  # Download the mirrored image

print("Image augmentation completed.")
```
